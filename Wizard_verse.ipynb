```python
from collections import Counter
import matplotlib.pyplot as plt
import PyPDF2
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob


# Function to read PDF
def read_pdf(pdf_path):
    pdf_text = ""
    with open(pdf_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            pdf_text += page.extract_text() + " "
    return pdf_text

# Function to clean text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)           # remove extra spaces/newlines
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) # remove punctuation
    return text

# Dictionary of PDF paths
pdf_files = {
    "HP1": "/content/HP1 - Harry Potter and the Sorcerer_s Stone.pdf",
    "HP2": "/content/HP2 - Harry Potter and the Chamber of Secrets.pdf",
    "HP3": "/content/HP3 - Harry Potter and the Prisoner of Azkaban.pdf",
    "HP4": "/content/HP4 - Harry Potter and the Goblet of Fire.pdf",
    "HP5": "/content/HP5 - Harry Potter and the Order of the Phoenix.pdf",
    "HP6": "/content/HP6 - Harry Potter and the Half-Blood Prince.pdf",
    "HP7": "/content/HP7 - Harry Potter and the Deathly Hallows.pdf"
}

# Read & clean all novels
novels_text = {}
novels_clean = {}

for key, path in pdf_files.items():
    print(f"Reading {key} ...")
    raw_text = read_pdf(path)
    novels_text[key] = raw_text
    novels_clean[key] = clean_text(raw_text)


eda_results = {}

for name, text in novels_clean.items():
    words = text.split()
    num_words = len(words)
    num_sentences = len(re.split(r'[.!?]', text))
    sentiment = TextBlob(text[:50000]).sentiment.polarity  # sample portion for speed
    common_words = [w for w in words if w not in stopwords.words('english')]
    top_words = Counter(common_words).most_common(15)

    eda_results[name] = {
        "Total Words": num_words,
        "Total Sentences": num_sentences,
        "Avg Sentence Length": round(num_words/num_sentences, 2),
        "Sample Sentiment": round(sentiment, 3),
        "Top Words": top_words
    }

eda_df = pd.DataFrame(eda_results).T
display(eda_df)
```

```python
import pandas as pd
from textblob import TextBlob
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load movie dialogues
dialogues = pd.read_csv('/content/Dialogue.csv', encoding='latin-1')

# Ensure proper column names exist
print("Dialogue columns:", dialogues.columns.tolist())

# Some files have 'movie', some 'Movie' column names
movie_col = 'movie' if 'movie' in dialogues.columns else 'Movies'
dialogue_col = 'dialogue' if 'dialogue' in dialogues.columns else 'Dialogue'
```

```python
#  2. Sentiment per movie
def movie_sentiment(group):
    text = ' '.join(str(x) for x in group if isinstance(x, str))
    return TextBlob(text[:50000]).sentiment.polarity  # limit for speed

movie_sentiments = dialogues.groupby('Chapter ID')[dialogue_col].apply(movie_sentiment).reset_index()
movie_sentiments.columns = ['Chapter ID', 'Sentiment']
```

```python
novel_sentiments = []
for name, text in novels_clean.items():
    sample = text[:50000]  # sample portion for speed
    polarity = TextBlob(sample).sentiment.polarity
    novel_sentiments.append({'Novel': name, 'Novel_Sentiment': polarity})

novel_sent_df = pd.DataFrame(novel_sentiments)
```

```python
# Try to align by HP1–HP7 order
comparison_df = pd.DataFrame({
    'Series': ['HP1','HP2','HP3','HP4','HP5','HP6','HP7'],
    'Novel_Sentiment': novel_sent_df['Novel_Sentiment'],
    'Movie_Sentiment': movie_sentiments['Sentiment'].iloc[:7].values
})
print("\n📊 Sentiment Comparison (first 7):")
display(comparison_df)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=comparison_df.melt(id_vars='Series', var_name='Type', value_name='Sentiment'),
            x='Series', y='Sentiment', hue='Type', palette='coolwarm')
plt.title("Harry Potter – Sentiment Comparison (Novels vs Movies)")
plt.ylabel("Average Sentiment Polarity")
plt.show()
```

```python
def top_words(text, n=20):
    words = re.findall(r'\b[a-zA-Z]{3,}\b', text.lower())
    return [w for w,_ in Counter(words).most_common(n)]

# Load the Chapters and Movies dataframes to link Dialogue to Movie Titles
chapters_df = pd.read_csv('/content/Chapters.csv', encoding='latin-1')
movies_df = pd.read_csv('/content/Movies.csv', encoding='latin-1')

# Merge dialogues with chapters to get Movie ID
dialogues_with_movie_id = pd.merge(dialogues, chapters_df[['Chapter ID', 'Movie ID']], on='Chapter ID')

# Merge with movies to get Movie Title
dialogues_with_movie_title = pd.merge(dialogues_with_movie_id, movies_df[['ï»¿Movie ID', 'Movie Title']], left_on='Movie ID', right_on='ï»¿Movie ID')

# Filter dialogues for HP1 (Harry Potter and the Sorcerer's Stone)
hp1_movie_dialogues = dialogues_with_movie_title[
    dialogues_with_movie_title['Movie Title'].str.contains("Sorcerer", case=False)
]['Dialogue'].astype(str)

movie_text_hp1 = ' '.join(hp1_movie_dialogues)
book_text_hp1 = novels_clean['HP1']

top_movie_words = top_words(movie_text_hp1, 20)
top_book_words = top_words(book_text_hp1, 20)
overlap = set(top_movie_words) & set(top_book_words)

print(f"\n🔍 Top Movie Words (HP1): {top_movie_words}")
print(f"🔍 Top Book Words (HP1): {top_book_words}")
print(f"💫 Common Themes (HP1): {overlap}")
```

```python

!pip install pytrends
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq


# 1. Setup Google Trends
pytrends = TrendReq(hl='en-US', tz=360)

# Keywords to track
keywords = ['Harry Potter', 'Hogwarts', 'Voldemort']

# Build payload for timeframe 2001-01-01 to 2025-12-31
pytrends.build_payload(kw_list=keywords, timeframe='2001-01-01 2025-12-31')

# Fetch interest over time
trend_df = pytrends.interest_over_time()

# Drop 'isPartial' column if present
if 'isPartial' in trend_df.columns:
    trend_df = trend_df.drop(columns=['isPartial'])

# ----------------------------
# 2. Plot trends over time
plt.figure(figsize=(12,6))
for keyword in keywords:
    plt.plot(trend_df.index, trend_df[keyword], label=keyword)
plt.title("Harry Potter Global Search Interest (2001–2025)")
plt.xlabel("Year")
plt.ylabel("Search Interest")
plt.legend()
plt.show()

# ----------------------------
# 3. Optional: Merge with movie release info
movies_meta = pd.read_csv('/content/Movies.csv')  # adjust path
# Ensure it has 'title' and 'release_year' columns
movies_meta['Release Year'] = pd.to_datetime(movies_meta['Release Year'], errors='coerce').dt.year

plt.figure(figsize=(12,6))
for keyword in keywords:
    plt.plot(trend_df.index, trend_df[keyword], label=keyword)
# Mark movie releases as vertical lines
for yr in movies_meta['Release Year'].dropna().unique():
    plt.axvline(pd.to_datetime(f"{int(yr)}-01-01"), color='gold', linestyle='--', alpha=0.6)
plt.title("Harry Potter Search Interest with Movie Releases")
plt.xlabel("Year")
plt.ylabel("Search Interest")
plt.legend()
plt.show()
```
